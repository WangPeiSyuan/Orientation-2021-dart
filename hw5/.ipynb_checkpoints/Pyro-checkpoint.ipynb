```python
import torch
import pyro
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('rugged_data.csv',  encoding='unicode_escape')
df = df[['rugged', 'cont_africa', 'rgdppc_2000']]
df = df.dropna(subset=['rgdppc_2000'])
af = df[df['cont_africa']==1].drop('cont_africa',axis=1)
naf = df[df['cont_africa']==0].drop('cont_africa',axis=1)
# print(af.describe())
# print(naf.describe())
af.hist(figsize=(10, 4))
naf.hist(figsize=(10, 4))

# print(df.shape)
# print(af.shape)
# print(naf.shape)
```

```python
## log transform
af[["rgdppc_2000"]] =  np.log(af[["rgdppc_2000"]])
naf[["rgdppc_2000"]] =  np.log(naf[["rgdppc_2000"]])
# print(af.describe())
af.hist(figsize=(10,4))
naf.hist(figsize=(10, 4))
```

```python
aX, ay = af['rugged'].to_numpy(), af['rgdppc_2000'].to_numpy()
nX, ny = naf['rugged'].to_numpy(), naf['rgdppc_2000'].to_numpy()
print(len(aX), len(ay))
print(len(nX), len(ny))
```

```python
from sklearn import datasets
from pyro.nn import PyroModule
from torch import nn

aX = torch.tensor(aX[:, np.newaxis], dtype = torch.float)
ay = torch.tensor(ay, dtype = torch.float)
nX = torch.tensor(nX[:, np.newaxis], dtype = torch.float)
ny = torch.tensor(ny, dtype = torch.float)
```

```python
def train(X, y):
    linear_reg_model = PyroModule[torch.nn.Linear](1,1)
    loss_fn = torch.nn.MSELoss(reduction='sum')
    optim = torch.optim.Adam(linear_reg_model.parameters(), lr=0.05)
    for i in range(1000):
        y_pred = linear_reg_model(X).squeeze(-1)
        loss = loss_fn(y_pred, y)
        optim.zero_grad()
        loss.backward()
        optim.step()
        if(i+1)%50==0:
            print('[epoch %d] loss: %.4f' % (i+1, loss.item()))
    return linear_reg_model
```

```python
linear_reg_model1 = train(aX, ay)
```

```python
linear_reg_model2 = train(nX, ny)
```

```python
for name, param in linear_reg_model1.named_parameters():
    print(name, param)
for name, param in linear_reg_model2.named_parameters():
    print(name, param)
```

```python
def plot(X, y, pred, title):
    plt.plot(X, y, "o")
    plt.plot(X, pred)
    plt.title(title)
    plt.xlabel("rugged")
    plt.ylabel("GDP")
    plt.legend()
```

```python
apred = linear_reg_model1(aX).detach().numpy()
plot(aX, ay, apred, "Africa")
```

```python
npred = linear_reg_model2(nX).detach().numpy()
plot(nX, ny, npred, "non Africa")
```
